**Title: Exploring and Visualizing Data in Python**

```python
#from google.colab import drive
#drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('/content/bike_buyers.csv')
print(df.head())
```

```python
 df.info()
 print("length of the dataset",len(df))
```




 **Data Preprocessing**

  **Identify and handling missing *

```python
print(df.isnull().sum())
```

```python
df['Income'].fillna(df['Income'].mean(),inplace=True)
df['Children'].fillna(df['Children'].mean(),inplace=True)
df['Cars'].fillna(df['Cars'].mean(),inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

Eliminate Missing values and Duplicates

```python
df=df.dropna()
print("Total rows after eliminate missing values:",len(df))
print(df.isnull().sum())
```

```python
df=df.drop_duplicates()
print(len(df))
```

```python
df.describe()
```

Data Visualization

```python
x=df['Age']
y=df['Income']
plt.scatter(x,y)
#plt.grid()
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Scatter Plot")
plt.show()
```

```python
sns.set(style='ticks')
sns.scatterplot(data=df,x='Age',y='Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Income by age- Scatter plot")
plt.show()
```

*From the above plot we can get that people age from 40 to 60 are making high average income*

```python
sns.set(style='ticks')
sns.barplot(data=df,x='Education',y='Income')
plt.xlabel('Education')
plt.ylabel('Income')
plt.title("Income by Education")
plt.xticks(rotation=45)
plt.show()
```

*From the above bar chart we can see that graduarte degree holder are making high income compared other education groups*

```python
sns.set(style='ticks')
sns.histplot(data=df,x='Income',bins=10)
bin_counts=df['Income'].value_counts(bins=10,sort=False)
for i,count in enumerate(bin_counts):
  plt.annotate(str(count),xy=(bin_counts.index[i].left,count),ha='center',va='bottom')

plt.title("Income Distribution")
plt.show()
```

*From the histogram , we can have clear picture of income distribution of people*

```python
sns.set(style='ticks')
sns.boxplot(data=df,y='Income')

plt.ylabel('Income')
plt.title("Income distribution")
plt.show()
```

*From the box plot, we can have clear idea that majority people making salary between 30 K to 70 K*

```python
plt.pie(df['Education'].value_counts(),labels=df['Education'].unique(),autopct='%1.2f%%')
plt.title("Education distribution")
plt.show()
```

*From the above pie chart , we can clearly get the insight that bachelors are majority group purchasing bike (30.16 %)*

```python
sns.violinplot(data=df,x='Education',y='Income')
plt.title("Income distribution")
plt.xticks(rotation=45)
plt.show()
```

**From the above pie chart , we can clearly get the insight that each group avergae income **

```python
sns.swarmplot(data=df,x='Education',y='Income')
plt.title("Income distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['Income','Age','Children']].corr()
sns.heatmap(correlation_matrix,annot= True)
plt.title("Correlation Matrix")
plt.show()
```

From the above heat map,
1.age and children are highly positively correlation with each other
2.Income and children are medium positively correlation with each other
3.Income and age are less positively correlation with each other

```python
correlation_matrix = df[['Income','Age','Cars']].corr()
sns.heatmap(correlation_matrix,annot= True)
plt.title("Correlation Matrix")
plt.show()
```

From the above heat map,
1.Income and cars are highly positively correlation with each other
2.Income and children are less positively correlation with each other
3.Cars and age are less positively correlation with each other
